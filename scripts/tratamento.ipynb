## Tratamento dos Dados

### Entendimento Inicial

#### Importanto dados e bibliotecas

```python
import pandas as pd

df = pd.read_csv("../data/raw/glp-2023-01.csv", delimiter=";")
df.head()
```

#### Obtendo uma visão geral da composição do dataset

```python
df.info()
```

#### Removendo coluna totalmente composta de valores nulos

```python
df = df.drop(columns=['Valor de Compra'])
df.head()
```

#### Atribuindo Tipos corretos onde necessário

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',', '.')
df['Valor de Venda'] = df['Valor de Venda'].astype(float)

df['Data da Coleta'] = pd.to_datetime(df['Data da Coleta'], dayfirst=True)
```

```python
df.info()
```

### Aplicando a mesma lógica para o restante dos datasets

### Importando datasets

```python
df1 = pd.read_csv('../data/raw/glp-2023-02.csv', delimiter=';')
df2 = pd.read_csv('../data/raw/glp-2024-01.csv', delimiter=';')
df3 = pd.read_csv('../data/raw/glp-2024-02.csv', delimiter=';')
df4 = pd.read_csv('../data/raw/glp-2025-01.csv', delimiter=';')
df5 = pd.read_csv('../data/raw/glp-2025-02.csv', delimiter=';')
```

### Tratando erros e valores faltantes

```python
def drop_coluna(dframe):
    dframe = dframe.drop(columns=['Valor de Compra'])
    return dframe


df1 = drop_coluna(df1)
df2 = drop_coluna(df2)
df3 = drop_coluna(df3)
df4 = drop_coluna(df4)
df5 = drop_coluna(df5)
```

```python
df2.head()
```

```python
def type_change(dframe):
    dframe['Valor de Venda'] = dframe['Valor de Venda'].astype(str)
    dframe['Valor de Venda'] = dframe['Valor de Venda'].str.replace(',', '.')
    dframe['Valor de Venda'] = dframe['Valor de Venda'].astype(float)
    dframe['Data da Coleta'] = pd.to_datetime(dframe['Data da Coleta'], dayfirst=True)
    return dframe

def convert_to_string_columns(dframe, exclude_cols=['Data da Coleta', 'Valor de Venda']):
    for col in dframe.columns:
        if col not in exclude_cols:
            dframe[col] = dframe[col].astype(pd.StringDtype())
    return dframe

frames = [df,df1,df2,df3,df4,df5]

for i, dframe in enumerate(frames):
    dframe = type_change(dframe)
    dframe = convert_to_string_columns(dframe)
    frames[i] = dframe
```

### Salvando datasets no formato .parquet

```python
df.to_parquet('../data/processed/glp-2023-01-processed.parquet')
df1.to_parquet('../data/processed/glp-2023-02-processed.parquet')
df2.to_parquet('../data/processed/glp-2024-01-processed.parquet')
df3.to_parquet('../data/processed/glp-2024-02-processed.parquet')
df4.to_parquet('../data/processed/glp-2025-01-processed.parquet')
df5.to_parquet('../data/processed/glp-2025-02-processed.parquet')
```
